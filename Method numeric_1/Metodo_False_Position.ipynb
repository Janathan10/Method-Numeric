```python
import math 
# Creating the function F(X)
def F(x):
    p = np.array([1, -1, -2], dtype=float)

    return np.polyval(p, x)

# Creating the Falseposition method
def falseposition(F, a, b, numIter, tol = 1e-3):
    #numIter is the counter number of iteractions
    #tol is the tolerance
    x0 = np.empty(shape=numIter) #The points of acurrate the function 
    y0 = np.empty(shape=numIter) #The points of acurrate the function
    
    fa , fb,  = F(a) , F(b), # Atribuition each values in each function
    
    #Verify if is a available value 
    if fa*fb >=0:
        print("Error ")
        return None #stop the program
    
    for i in range(0, numIter): #The method is going to be apply the number of iteractins stabilish 
        x = (a*fb)-(b*fa)/(fb-fa) # the method
        fx  = F(x) # Atribuition the values in the function
        x0[i] = x # Update the points
        y0[i] = fx #Update the points
        #Verify which one of interval extremes must be update
        if fa*fx < 0:
            #Update the extreme b and f(b)
            b = x
            fb = fx
        if fb*fx < 0:
            #Update the extreme a and f(a)
            a = x
            fa = fx
        if abs(fx) <=tol : #Considering of tolerance on the function value
            print("the precision is accurate ")
            break ;
        
    return (x0, y0)
```

```python
# Apply the false position method
a, b = -0.9 ,2.1 # Stabilish the input values
numIter = 10  # Stabilish the number of iteractions
x0, y0 = falseposition(F, a, b, numIter ) #Call it the falseposition function
# Show the results
x = np.arange(a, b, np.abs(b-a)/100)
y = F(x)

# Display on screen the method apply
plt.figure()
plt.plot(x, y, 'b-')
plt.plot(x0, y0, 'ro')
plt.xlabel('$x$')
plt.ylabel('$p(x)$')
plt.grid(True)
for i in range(0, numIter):
    plt.text(x0[i], y0[i], i)
plt.show()

for i in range(0, numIter):
    print('Iteration: ', i, ' Approx. root: ', x0[i])
```
