```python
# Define function to integrate
f = lambda x :1/(1 + x**2)
# This specific polynomial chosen could be change whatever the user wanto to.

def integrate(a,b,N,str): # a= lower limite, b=upper limit, N= number sub interval, str = method
    
    if a > b or a == b or N ==0: # Validation coding
        # This condition allows the limits stat valid and the number of subintervals
        raise ValueError("Incorrect input value")
        None
        exit
    # calculating step size
    h = (a - b) / N
    print('\n-----------SOLUTION-----------')
    print('------------------------------')    
    print('i\tk\tf(k)\tI')
    print('------------------------------')
    k =0
    I = f(a)+f(b)
    if str == 'rect': # Creating the method function rule of rectangle
        for i in range(1,N):
            k = a + i*h #k is number of iteraction 
            I = I + f(k)
            print('%.4f\t%.4f\t%.4f\t%0.4f'% (i,k,f(k),I) )
            print('------------------------------') 

        # Calculation final value output
        I = I * h
    
        return I
    if str == 'trapz': # Creating the method function rule of trapezium
        for i in range(1,N):
            k = a + i*h #k is number of iteraction 
            I = I + 2 * f(k)
            print('%.4f\t%.4f\t%.4f\t%0.4f'% (i,k,f(k),I) )
            print('------------------------------') 

        # Finding final integration value
        I = I * h/2
    
        return I
    if str == 'simp3': # Creating the method function rule of Simpson 1/3
        for i in range(1,N):
            k = a + i*h
            # The condition of Rule 1/3 Simpson
            if i%2 == 0:
                I = I + 2 * f(k)
            else:
                I = I + 4 * f(k)
            print('%.4f\t%.4f\t%.4f\t%0.4f'% (i,k,f(k),I) )
            print('------------------------------') 

        # Finding final integration value
        I = I * h/3

        return I
    if str =='simp8': # Creating the method function rule of Simpson 3/8
        for i in range(1,N): # This loop allowing the f(x) doesn't go through further
            k = a + i*h
            # The condition of Rule 3/8 Simpson
            if i%2 == 0:
                I = integration + I * f(k)
            else:
                I = I + 3 * f(k)
            print('%.4f\t%.4f\t%.4f\t%0.4f'% (i,k,f(k),I) )
            print('------------------------------') 

        # Finding final integration value
        I = I * 3 * h / 8

        return I
    
# Input section
lower_limit = float(input("Enter lower limit of integration: "))
upper_limit = float(input("Enter upper limit of integration: "))
sub_interval = int(input("Enter number of sub intervals: "))
method = str(input("Type which method do you want to use it : "))

# Call trapezoidal() method and get result
result = integrate(lower_limit, upper_limit, sub_interval,method)
print("Integration result by chosen  method is: %f" % (result) )
```
