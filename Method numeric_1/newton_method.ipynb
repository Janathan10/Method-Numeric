```python
%matplotlib notebook
#Creating the function f(x) = ln(x)-1
def f(x):
    return np.log(x)-1

x = np.arange(-5, 20, 0.01)
y = f(x)

# Display on screen the function
plt.figure()
plt.plot(x, y, 'r-.')
plt.xlabel('$x$')
plt.ylabel('$p(x) = log(x)-1 $')
plt.grid(True)
plt.show()
```

```python
%matplotlib notebook
#Creating the fuction f(x)
def f(x):
    return np.log(x)-1
# Creating the derivative of function f(x)
def df(x):
    return 1/x

#Creating the method of newton
def Newton(f, df, x0 ,tol=1e-9, disp =False): # Input values
    iterNum =1; #Iniciating one unit of counter
    while True: #This loop going to happen until break the condition abs(x-x0) < tol
        x= x0 - f(x0)/df(x0)  # The method
        
        if disp: #Display on screen the Table of each iteractions
            print("Iter: %3d | x: %+.7e | f(x): %+.7e | f'(x): %+.7e| x: %+.7e" \
                  %(iterNum, x0, f(x0), df(x0), x))
        if abs(x-x0) < tol:
            break; #Break the code if the tolerance is achieve
        x0=x;
        iterNum +=1 # Increase one unit of counter number interactions
    return x

Xns = Newton(f, df, 2.7, disp= True)

#Plotting the graphic of f(x) for x in [2:3,4] and plotting the value of Xns as well.
x = np.arange( 2, 3.4 ,0.1)
y = f(x)
plt.figure()
plt.plot(x, y, 'r-.', label = '$f(x)$') 
plt.plot(Xns, f(Xns), 'bo', label = '$X_{NS}$')
plt.grid()
plt.ylabel('$f(x)= log(x)-1 $')
plt.xlabel('$x$')
plt.legend()
plt.show()
```
