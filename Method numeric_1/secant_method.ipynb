```python
# Populates the notebook with the most used numerical packages
%pylab notebook
# Import numpy for array processing
import numpy as np
# Import pyplot for graphics
import matplotlib.pyplot as plt

```

No método da Secante, foi usado como $f(x) = x^2 - e^x$, com valor de tolerância $ 1e^-2$, com os respectivos valores de $ x_0 e x_1 = 0.5 , 1 $

```python
# Showing the graphics of functions $f(x) = x^2 - e^x$
%matplotlib notebook
x = np.arange( -10, 20 ,0.01) #Creating the size of x to ilustration the graphics

#Creating the function x^2 - e^x
def f(x):
    return x**2 - np.exp(x)

#Display on screen the function  x^2 - e^x
y = f(x)
plt.figure()
plt.plot(x, y, 'r-.', label = '$f(x)$')
plt.ylabel('$f(x)= x^2 -e^x$')
plt.xlabel('$x$')
plt.legend()
plt.grid()
plt.show()
```

```python
# Creating the function Secant method
def Secant( f, x0 ,x1, tol=1e-9, disp=False): #stabilish the tolerance value
    iterNum =1; #Iniciating the counter variable
    while True:# This loop is going to happen until the condition  abs(x -x1) < tol breaks.
        x= x1-f(x1)*(x1-x0)/(f(x1)-f(x0)) #the method
        
        if disp: #Display on screen each iteractions 
            print("Iter :%3d| x0:%+.5e|  x1:%+.5e| f(x0):%+.5e | f(x1): %+.5e| x:%+.5e " \
           % (iterNum, x0, x1, f(x0), f(x1), x))
        if abs(x -x1) < tol :
            break;
        x0, x1 = x1, x;
        iterNum +=1 #Increase the input counter number of interactios
    return x   
```

```python
# Display each iteraction and the graphics
%matplotlib notebook
xNS = Secant(f, 0.5, 1, tol = 1e-2, disp= True) #This showing the number of iteractions until the tolerance is satisfied
xNS
x = np.arange( -5, 5 ,0.001) # imaginary size of x
#Plotting the graphic of f(x) for x in [-5, 5 ,0.001] and plotting the value of Xns as well.
y = f(x)
plt.figure()
plt.plot(x, y, 'r-.', label = '$f(x)$') 
plt.plot(xNS, f(xNS), 'bo', label = '$X_{NS}$')
plt.grid()
plt.ylabel('$f(x)= x^2-e^x $')
plt.xlabel('$x$')
plt.legend()
plt.show()
```
