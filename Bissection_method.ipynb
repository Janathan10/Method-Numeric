```python
# Organizing the coefficients of the polynomial
p = np.array([1, 12, -100, -6], dtype=float)

# The roots of the polynomial
r = np.roots(p)
print('Roots: ', r)

# Plot the function graph and marking the roots
x = np.arange(-30, 20, 0.01)
y = np.polyval(p, x)
x0 = r
y0 = np.polyval(p, r)

# Display on screen 

plt.plot(x, y, 'r')
plt.plot(x0, y0, 'bo')
plt.xlabel('$x$')
plt.ylabel('$p(x) = x^3 + 12x^2 - 100x - 6$')
plt.grid(True)
plt.show()
```

```python
# Creating the function F(X)
def F(x):
    p = np.array([1, 12, -100, -6], dtype=float)
    return np.polyval(p, x)

# Bisection method
def Bisection( F, a, b, numIter,tol = 1e-3):
    
    #numIter is the counter number of iteractions
    #tol is the tolerance
    x0 = np.empty(shape=numIter) #The points of acurrate the function 
    y0 = np.empty(shape=numIter) #The points of acurrate the function
    
    fa , fb,  = F(a) , F(b) # Atribuition each values in each function
    
    #Verify if is a available value 
    if fa*fb >=0:
        print("Error ")
        return None #stop the program
    
    for i in range(0, numIter): #The code going to compile number of time the iteractions stabilish
        x = (a+b)/2 # The method
        fx = F(x)
        x0[i] = x  # The input value
        y0[i] = fx
        #Verify which one of interval extremes must be update
        if fa*fx < 0:
            #Update the extreme b and f(b)
            b = x
            fb = fx
        if fb*fx < 0:
            #Update the extreme a and f(a)
            a = x
            fa = fx
        if abs(fx) <=tol : #Considering of tolerance on the function value
            print("the precision is accurate ")
            break ;  # the code is going to break when this condition come true
        
    return (x0, y0)

# Applying the BISECTION METHOD 
a, b = -18, -17 #Stabilish the input values
numIter = 5 # #Stabilish the input values of counter
x0, y0 = Bisection(F, a, b, numIter , tol= 1e-3 ) #Stabilish the number of t 1e-3
# Show the results
x = np.arange(a, b, np.abs(b-a)/1000)
y = F(x)

# Display on screen the method apply
plt.figure()
plt.plot(x, y, 'b-')
plt.plot(x0, y0, 'ro')
plt.xlabel('$x$')
plt.ylabel('$p(x)$')
plt.grid(True)
for i in range(0, numIter):
    plt.text(x0[i], y0[i], i)
plt.show()

for i in range(0, numIter):
    print('Iteration: ', i, ' Approx. root: ', x0[i])
```
