```python
import numpy as np
import math
import matplotlib.pyplot as plt 
%matplotlib notebook

def polynomial_regression(x: np.ndarray, y: np.ndarray, order):  #n number of higher exp
    # VALIDATION CODE
    if x.size !=y.size: #If the both vector isn't the sami size of dimension, the code should stop
        raise ValueError("ERROR, Incorrect input data ")
        None
        
    n = x.size #This variable is the number of element from the function
    
    Sx,Sy,Sx2,Sx3,Sx4,Sxy,Sx2y = np.sum(x),np.sum(y),np.sum(x**2),np.sum(x**3),np.sum(x**4),np.sum(x*y),np.sum((x**2)*y)
    ##Creating the matrix S*a= s , same idea Ax=b
    S = np.array([[n,Sx,Sx2], [Sx,Sx2,Sx3],[Sx2,Sx3,Sx4]])
    s = np.array([Sy,Sxy,Sx2y])
    a = np.linalg.inv(S)@s #vector product
    
    xx = np.arange(0,10 ,0.1)  # Creating the lenght of axis X 
    yy = a[0] 
    #This looping create the polynomial y for each interaction and adding exponent until the order
    for i in range(1,order,1):
        yy += a[i]*np.power(xx,i)
    
    print("S : ", S)
    print("s : ", s)
    for i in range (len(a)) :
        print("a"+str(i)+"=",a[i])
    print("Using polytfit : ", np.polyfit(x,y,order))
    
    plt.plot(x,y,'ro',xx,yy,'b-')
    plt.xlabel('$x$')
    plt.ylabel('$y$')
    plt.grid()
    plt.show()
```

```python
x = np.array([2,4,5,9,10])
y = np.array([2.2,3,4.4,5,5.5])
print("x:",x)
print("y:",y)
polynomial_regression(x,y,3)
```
